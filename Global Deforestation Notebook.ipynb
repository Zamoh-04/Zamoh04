## 4. Data Cleaning <a class="anchor" id="chapter4"></a>

<div class="alert alert-block alert-info">
<b>Data cleaning</b> is a crucial step in the data analysis process, involving the correction or removal of incorrect, corrupted, duplicate, or incomplete data within a dataset. Through various techniques such as filling missing values, removing outliers, and standardizing data formats, it ensures the accuracy and reliability of subsequent analyses and decision-making.
</div>


The `check_null_values` function iterates over each column in a DataFrame to identify and print the count of null values, providing column-specific null counts.

```python
def check_null_values(df_copy):
    """
    Print the count of null values for each column in a DataFrame.

    This function iterates through each column in the DataFrame to check for the presence of null values.
    If a column contains null values, it prints the column name along with the number of null values.

    Parameters:
    df (DataFrame): The pandas DataFrame to check for null values.

    Returns:
    None: This function does not return a value; it only prints information.
    """
    for column in df_copy:
        if df_copy[column].isnull().any():
            print('{0} has {1} null values'.format(column, df_copy[column].isnull().sum()))
```

```python
check_null_values(forest_area_km_df_copy)
```

```python
check_null_values(forest_area_percent_df_copy)
```

**Results** : The year 1990 has the highest number of null values. We will have to use the global mean to replace the null values for each year. 

The `count_duplicate_rows` function checks a DataFrame for duplicate entries by utilizing the duplicated method, , and then summarizes the total number of duplicate rows. This function is useful for data cleaning processes, ensuring uniqueness in datasets by returning the count of duplicate rows.

```python
def count_duplicate_rows(df_copy):
    """
    Count the number of duplicate rows in a DataFrame.

    This function calculates the total number of duplicate rows in the DataFrame by calling the `duplicated` method,
    which marks duplicates as `True`, and then sums these cases.

    Parameters:
    df_copy (pandas.DataFrame): The DataFrame to check for duplicates.

    Returns:
    int: The count of duplicate rows.
    """
    duplicate_count = df_copy.duplicated().sum()
    return duplicate_count
```

```python
count_duplicate_rows(forest_area_percent_df_copy)
```

```python
count_duplicate_rows(forest_area_km_df_copy)
```

**Results** : There are no duplicate entries. We don't need to delete any rows. 

The `preprocess_data` function performs essential data cleaning operations on a DataFrame, setting missing values, filling in gaps with statistical measures (mean and median), and dropping non-essential columns. After processing, it prints the first two rows of the cleaned DataFrame for quick inspection and returns the entire modified DataFrame for further use.


```python
def preprocess_data(df_copy):
    """
    Perform basic data cleaning on the DataFrame.

    This includes:
    - Setting missing 'seasons' to 0.
    - Filling missing IMDb scores and votes with the mean and median, respectively.
    - Filling missing TMDB popularity and scores with the median and mean, respectively.
    - Dropping columns with many null values or that are not required.


    Returns:
        DataFrame: The cleaned DataFrame.
    """
    
    # Assign zero seasons to movies where the number of seasons is missing
    df_copy['seasons'] = df_copy['seasons'].fillna(0)
    
    # Fill missing IMDb scores with the mean score
    df_copy['imdb_score'] = df_copy['imdb_score'].fillna(df_copy['imdb_score'].mean())
    
    # Fill missing IMDb votes with the median number of votes
    df_copy['imdb_votes'] = df_copy['imdb_votes'].fillna(df_copy['imdb_votes'].median())
    
    # Fill missing TMDB popularity scores with the median popularity
    df_copy['tmdb_popularity'] = df_copy['tmdb_popularity'].fillna(df_copy['tmdb_popularity'].median())
    
    # Fill missing TMDB scores with the mean score
    df_copy['tmdb_score'] = df_copy['tmdb_score'].fillna(df_copy['tmdb_score'].mean())
    
    # Drop columns that are not required or have many null values
    df_copy = df_copy.drop(['imdb_id', 'description', 'age_certification'], axis=1)
    
    return df_copy
```

```python
# look at the top 2
preprocess_data(forest_area_percent_df_copy.head())
```

## 5. Exploratory Data Analysis (EDA) <a class="anchor" id="chapter5"></a>


**Exploratory Data Analysis (EDA)** is a pivotal step in data analysis, focusing on understanding the main characteristics of a dataset through visual and statistical exploration. It involves summarizing the dataset's size, structure, and basic statistics, while employing various visualization techniques like histograms, scatter plots, and box plots to uncover patterns, relationships, and anomalies within the data. By iteratively exploring the data, researchers generate hypotheses, identify trends, and refine their understanding of the dataset, thereby laying the groundwork for subsequent analyses and insights.
 

---

Filtering the DataFrame allows for the examination of various deforestation trends based on different criteria such as country name, income group, and region. By filtering the data by country name, it becomes possible to analyze deforestation trends within specific nations, offering insights into localized environmental changes and conservation efforts. Exploring deforestation trends across income groups enables the identification of disparities in forest management practices and environmental impacts between low-income, middle-income, and high-income countries, highlighting the influence of socioeconomic factors. Filtering by region allows for a broader geographical analysis of deforestation patterns, facilitating the understanding of regional differences in forest cover changes and the underlying drivers of deforestation across different parts of the world. Through such filtering, it becomes possible to gain comprehensive insights into deforestation trends and their drivers, informing targeted conservation strategies and policy interventions aimed at mitigating forest loss and promoting sustainable land management practices.

```python
# Create a list of country names from index 0 to 209
country_names_to_include = forest_area_percent_df.index[:210].tolist()

# Filter the DataFrame to include only rows where the "Country Name" matches the specified list of country names
filtered_country_names_df = forest_area_percent_df[forest_area_percent_df.index.isin(country_names_to_include)]

# Display the filtered DataFrame
filtered_country_names_df.head()
```

```python
income_groups_to_include = [
    'High income',
    'Low income',
    'Upper middle income',
    'Lower middle income',
    'Middle income',
]

# Filter the DataFrame to include only rows where the "Country Name" matches the specified income groups
filtered_income_group_df_copy = forest_area_percent_df_copy[forest_area_percent_df_copy['Country Name'].isin(income_groups_to_include)]

# Display the filtered DataFrame
filtered_income_group_df_copy.head()
```

```python
regions_to_include = ['East Asia & Pacific',
'Europe & Central Asia',
'Latin America & Caribbean',
'Middle East & North Africa',
'North America',
'South Asia',
'Sub-Saharan Africa',
'Central Europe and the Baltics',
'Africa Eastern and Southern'
]
filtered_by_region_df = forest_area_percent_df_copy[forest_area_percent_df_copy['Country Name'].isin(regions_to_include)]

# Display the filtered DataFrame
filtered_by_region_df.head(7)
```

## Visual Representations of the Filtered Data

```python
mean_by_income_group = filtered_income_group_df_copy.groupby('Country Name').mean(numeric_only=True)
mean_by_income_group.plot(kind='line', figsize=(16, 6))

# Increase the font size of the title
plt.title('Mean Deforestation Rate by Income Group', fontsize=20)

plt.xlabel('Income Group')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.tight_layout()

# Move the legend to the upper right corner
plt.legend(loc='upper right')

plt.show()
```

**Results** :

This analysis of deforestation trends over time is made using mean deforestation rates. The mean deforestation rate was chosen to capture the average trend while smoothing out short-term fluctuations. By focusing on the mean, the analysis aimed to discern overall patterns in deforestation rates, aiding in the identification of periods of change. Comparisons between different time periods were facilitated, offering insights into the evolution of deforestation rates over the years. Ultimately, this approach provides valuable insights for informing conservation efforts and policy interventions aimed at addressing deforestation.

---

#### This plot `Forest Area Over Time For Each Region` shows how the forest area has changed over time for different Regions

```python
plt.figure(figsize=(10, 6))

# Iterate over the filtered DataFrame rows
for index, row in filtered_by_region_df.iterrows():
    # Extract region name
    region_name = row['Country Name']
    
    # Extract forest area data for the region (starting from column '1990')
    forest_area = row.iloc[2:]  # Exclude the first two columns (Country Name and Country Code)
    
    # Plot the data as a line chart
    plt.plot(forest_area.index, forest_area, label=region_name)

# Add labels and title
plt.title('Forest Area Over Time for Each Region')
plt.xlabel('Year')
plt.ylabel('Forest Area (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Add legend outside the plot

# Show the plot
plt.show()
```

**Results** :

A line graph showing forest area changes over time per region visually presents how forest coverage has evolved across different areas. It reveals changes in forested land within each region, highlighting trends like deforestation, reforestation, and stable forest areas. We can observe these trends in regions such as Sub-Saharan Africa and Latin America & Caribbean where they have seen a slight decrease in forest area, while regions like North America where saw an increase in forest area.There are other regions like Middle East & North Africa and Europe & Central Asia that show stability in forest area throughout the years, which means there has not been much change in their forest area. This can help  policymakers, environmentalists, and researchers spot patterns and differences in how forests are managed and in making informed decisions on conservation, land-use policies, and sustainable resource management tailored to each region's specific needs.
